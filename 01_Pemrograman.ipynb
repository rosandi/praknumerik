## Penulisan badan program

Badan program pada python ditulis dalam teks (_plain text_) di kolom pertama. Setiap baris diakhiri dengan karakter garis baru (`'\n'`).
Baris kode yang ditulis pada kolom pertama adalah bagian dari badan program utama. Blok program ditandai dengan indentasi, dapat dengan tabulasi (`\t`) atau spasi, dan harus konsisten dalam seluruh badan program. Jumlah tabulasi dan spasi harus konsisten dalam satu blok.

## Variabel

Python menggunakan tipe data otomatis pada variabel. Ini berarti ketika variabel diinisialisasi, tipe variabel menyesuaikan sesuai dengan data yang diberikan. Nama dari variabel diawali oleh huruf non-bilangan atau garis bawah (_underscore_) tidak mengandung spasi dan karakter-karakter khusus. Untuk menampilkan data ke layar digunakan fungsi `print()`.

````python
# Contoh membuat variabel

nama="Agus Tine"
usia=23
Tinggi=1.7
info=f"Nama: {nama}\nUsia: {usia}\nTinggi badan: {Tinggi}"

print(info)

type(usia)
print("tipe nama:",type(nama))
print("tipe usia:",type(usia))
print("tipe Tinggi:",type(Tinggi))
print("tipe info:",type(info))
````

Dalam Python variabel yang telah digunakan dapat digunakan kembali dengan tipe data yang berbeda.

````python
type(usia)
print("tipe usia:",type(usia))
usia="Dua puluh tiga tahun"
print("tipe usia:",type(usia))

print()
print(info)

print()
info=f"Nama: {nama}\nUsia: {usia}\nTinggi badan: {Tinggi}"
print(info)
````

## String

String dalam Python dituliskan diantara tanda kutif. Tidak ada perbedaan tanda kutif tunggal dan ganda. Dalam string dapat mengandung karakter _escape_ seperti `\n` untuk garis baru, `\t` untuk tabulasi.

````python
print('Data ini\n        diatas\rbaris\nmenggunakan\ttabulasi\tseragam')
````

Variabel dapat disubstitusi ke dalam string secara terformat. 

````python
a=3
b=2.134
c=5/7

print("data tidak diformat:")
print("tipe a:", type(a))
print("tipe b:", type(b))
print("tipe c:", type(c))

print()
print("a=",a)
print("b=",b)
print("c=",c)

print()
print("dalam sebuah string:")
print("a={}\nb={}\nc={}".format(a,b,c))

print()
print("mencetak dengan format:")
print("a=%d"%a)
print("b=%0.3f"%b)
print("c=%0.2e"%c)

print()
print("dengan menggunakan 'f-string'")
print(f"a={a}\nb={b}\nc={c}")
print("dengan menggunakan 'f-string' berformat")
print(f"a={a:10}\nb={b:0.2f}\nc={c:0.2e}")
````

## Operator

Berikut operator yang dapat digunakan untukk perhitungan aritmetika.

$$
\begin{array}{|c|l|}
\hline
\text{+} & \text{penjumlahan} \\ \hline
\text{-} & \text{pengurangan} \\ \hline
\text{*} & \text{perkalian} \\ \hline
\text{/} & \text{pembagian} \\ \hline
\text{//} & \text{pembagian integer} \\ \hline
\text{**} & \text{pangkat} \\ \hline
\text{\%} & \text{modulo, sisa pembagian} \\ \hline
\end{array}
$$

Operator memiliki hirarki yang menentukan operasi mana yang dikerjakan terlebih dahulu. Hirarki paling tinggi adalah pangkat, kemudian perkalian, pembagian, dan modulo. Hirarki paling akhir adalah penjumlahan dan pengurangan. Agar dalam ekspresi matematika suatu operasi dikerjakan paling dulu, harus ditulis di dalam kurung `()`.

````python
a1=1+2*7+4/2+1
a2=(1+2)*7+4/(2+1)

print(f'nilai a1={a1:0.2f}\nnilai a2={a2:0.2f}')
````

**Operator augmentasi**
Operator augmentasi menunjukkan operasi aritmetika yang hasilnya di simpan di dalam variabel yang dioperasikan. Misalnya untuk variabel bernama `a` dapat dilakukan seperti pada tabel berikut.

$$
\begin{array}{|l|l|}
\hline
\text{ekspresi} & \text{arti} \\ \hline
\text{a+=1}  & \text{a=a+1} \\ \hline
\text{a-=2}  & \text{a=a-1} \\ \hline
\text{a*=10} & \text{a=a*10} \\ \hline
\text{a/=2}  & \text{a=a/2} \\ \hline
\text{a**=3} & \text{a=a**2} \\ \hline
\text{a\%=2} & \text{a+a\%2} \\ \hline
\end{array}
$$

**Operator relasi**

$$
\begin{array}{|c|l|}
\hline
\text{<}  & \text{kurang dari} \\ \hline
\text{>}  & \text{lebih dari} \\ \hline
\text{<=} & \text{kurang dari atau sama dengan} \\ \hline
\text{>=}  & \text{lebih dari atau sama dengan} \\ \hline
\text{==} & \text{sama dengan} \\ \hline
\text{!=} & \text{tidak sama dengan} \\ \hline
\end{array}
$$


## Jenis data terstruktur

**List**

Sebuah list adalah deret dari banyak data. Tipe data dalam list boleh beragam.

```python
alis=[1, 2, 'a', 'g', "ini anggota terakhir"]
```

Isi dari sebuah list diakses melalui indeks yang dimulai dengan $0$.

**Tuple**

Tuple adalah data berisi konstanta, yang tidak dapat diubah satu persatu.

```python
loc=(106.889, 2.334)
tier=("cat","fish")
```

Isi dari tuple diakses sama dengan list menggunakan indeks mulai dari $0$, akan tetapi tidak dapat diubah secara individual.

**Dictionary**

Dictionary adalah data yang memiliki anggota yang memiliki kunci sebagai penunjuk.

```python
data={
    "nama": "Agus Tine", 
    "umur": 23, 
    "Tinggi": 1.7, 
    "daftar_teman": ["Jack", "John", "George"],
    "teman": ("cat", "fish")
}
```

Anggota dari dictionari diakses melalui nama kuncinya, dan isinya boleh diubah.

# Pengulangan

Dalam Python pengulangan dapat menggunakan kata kunci `while` dan `for`.

Menggunakan `while`:

```python
while kondisi:
    print("dijalankan selama kondisi benar")
```

Menggunakan `for`:

```python
for i in deret:
    print("dijalankan untuk setiap anggota deret")
```

`deret` dapat berupa data banyak atau menggunakan fungsi `range()`.

````python
# Bekerja dengan list
daftar_buah = [
    "Apel", 
    "Jeruk", 
    "Pisang", 
    "Mangga", 
    "Anggur", 
    "Semangka", 
    "Melon", 
    "Nanas", 
    "Pepaya", 
    "Stroberi", 
    "Durian", 
    "Rambutan", 
    "Manggis", 
    "Salak", 
    "Alpukat"
]

print(daftar_buah)
````

````python

# Mengambil rata-rata jumlah huruf dalam nama buah

hrata=0
for i in range(len(daftar_buah)):
    hrata+=len(daftar_buah[i])
    
hrata/=len(daftar_buah)
print(f'rata-rata jumlah huruf dalam nama-nama buah: {hrata}')
````

````python
# Mengambil hanya buah berawalan huruf A
abuah=[]

for buah in daftar_buah:
    if buah.startswith('A'):
        print(f"buah {buah} diambil")
        abuah.append(buah)
    else:
        print(f'buah {buah} ditolak')

print('ini adalah daftar nama buah diawali huruf "a"')
print(abuah)
````

````python
# Iterasi list menggunakan while
i=0
while i<len(daftar_buah):
    print(daftar_buah[i])
    i+=1
````

## Fungsi

Fungsi adalah bagian dari program yang ditulis sebagai kesatuan sehingga dapat dipanggil berkali-kali tanpa menulis ulang bagian badah program. Sebuh fungsi dapat menerima argumen dan mengembalikan nilai,

## Membaca dan menulis file

Data dalam Python dapat ditulis dengan mudah dalam bentuk teks. File dapat berisi data berformat atau hanya teks mentah. File ini memiliki isi yang identik dengan tampilan layar.


````python
with open("databuah.txt","w") as fdata:
    fdata.write("nomor\tnama\t berat\n")
    for i in range(len(berat_buah)):
        fdata.write(f'{i}\t{berat_buah[i][0]}\t{berat_buah[i][1]}\n')

print("file databuah.txt sudah dibuat")
````

````python
# Menggunakan format
with open("databuah.txt","w") as fdata:
    fdata.write("nomor     nama  berat\n")
    for i in range(len(berat_buah)):
        fdata.write(f'{i:4}{berat_buah[i][0]:>10}{berat_buah[i][1]:5}\n')

print("file databuah.txt sudah dibuat")
````

````python
!cat databuah.txt
````

**Membaca file**

````python
with open("databuah.txt") as fdata:
    baris=fdata.readlines()

print(baris)
````

````python
# menampilkan isi file sesuai baris
for ln in baris:
    print(ln.strip())
````

````python
# menterjemahkan baris (parsing)
nomor=[]
nama=[]
berat=[]

judul=True

for ln in baris:
    
    if judul: 
        judul=False
        continue
    ln=ln.strip()
    ln=ln.split(' ')
    dd=[]
    for it in ln:
        if it == '': continue
        dd.append(it)
        
    print(dd)
    nomor.append(dd[0])
    nama.append(dd[1])
    berat.append(dd[2])

print()
for i in range(len(nomor)):
    print(f'{nomor[i]} {nama[i]} {berat[i]}')

print()
data={
    "nama": nama,
    "berat": [eval(a) for a in berat]
}
print(data)
````

````python
# Menuliskan data dalam CSV
with open('databuah.csv','w') as fcsv:
    fcsv.write('no,nama,berat\n')
    for i in range(len(nomor)):
        ln=",".join([nomor[i], nama[i], berat[i]])
        fcsv.write(f'{ln}\n')

print('file databuah.csv sudah dibuat')
````

## Menggunakan pustaka Numpy

Python memiliki pustaka standard untuk bekerja dengan numerik bernama **Numpy**. Pustaka ini memiliki struktur data yang digunakan khusus untuk array numerik, berupa `numpy.array`. Semua fungsi-fungsi matematika dalam **Numpy** dapat bekerja dengan data array.

Selain `numpy` kita akan sering melakukan plot data. Untuk pekerjaan ini anda akan menggunakan pustaka `matplotlib`.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# ubah berat ke array numerik
nberat=[eval(b) for b in berat]
plt.plot(nberat,'o')
plt.plot(nberat,'-')
plt.xlabel('jenis buah')
plt.ylabel('gram')
plt.show()
````

````python
# data buah dengan kalori
import csv

with open('buahkalori.csv') as cfile:
    clines=csv.reader(cfile)
    nr=[]
    nm=[]
    ms=[]
    cal=[]
    i=0
    for ln in clines:
        if i==0:
            i+=1
            continue
        nr.append(i)
        nm.append(ln[0])
        ms.append(float(ln[1]))
        cal.append(float(ln[2]))
        i+=1

for i in range(len(nr)):
    print(f'{nr[i]} {nm[i]} {ms[i]:0.2f} {cal[i]:0.2f}')
````

````python
plt.plot(nr,ms)
plt.plot(nr,cal)
plt.xticks(nr,nm,rotation='vertical')
plt.show()
````
