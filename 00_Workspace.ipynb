## Petunjuk penggunaan _Docker Container_

Dalam sistem operasi modern kita dapat membangun sebuah sistem untuk bekerja yang terisolasi dari komputer yang digunakan. Cara bekerja seperti ini disebut dengan _kontainerisasi_, yang memungkinkan untuk membangun sebuah sistem lengkap dalam satu image sistem operasi. Teknik komputasi seperti ini memberikan keleluasaan bagi pengembang untuk merancang dan mengimplementasikan software tanpa harus berbentrokan dengan sistem operasi utama yang digunakan. Semua keterkaitan perangkat lunak terdapat secara lengkap dalam sebuah kontainer.

Untuk kuliah Metode Numerik telah disiapkan citra sistem operasi dalam bantuk docker image. Setelah instalasi Docker desktop selesai, sebuah image kontainer docker dapat ditarik (pull) dari repositori online. Cara mudah untuk menjalankan kontainer adalah menggunakan ``compose.yml`` file yang diaktifkan dengan _docker compose_. Dalam file ini sudah terdapat seting-seting yang penting, sehingga tidak perlu dituliskan langsung di baris perintah. Berikut adalah isi dari file ``compose.yml`` yang memiliki format file text yang bernama _yaml_.

---

```yaml
services:
  numeric:
    image: rosandi/nrxalp
    hostname: numeric
    container_name: nrx
    ports:
      - 8888:8888
    volumes:
      - ./:/workdata
```

---

Tahap-tahap yang harus dilakukan adalah sebagai berikut,

1. Instal **Docker Desktop** dari laman `https://www.docker.com` untuk sistem operasi yang ada gunakan.
2. Buat file _yaml_ diatas dan simpan di direktori tempat kerja.
3. Di terminal atau _Powershell_ untuk sistem Windows pindah ke direktori tempat kerja.
4. Tarik image docker dan jalankan menggunakan perintah \
   `docker compose up` \
   Docker akan mengambil image dari repositori jika belum ada di mesin lokal.
6. Dengan menggunakan browser buka laman `http://localhost:8888`
7. Lingkungan kerja untuk kuliah ini sudah siap dan dapat digunakan melalui sistem web yang bernama _Jupyter Notebook_.

## Berinteraksi dengan sistem komputer melalui Shell

Sebuah _shell_ adalah antarmuka antar pengguna dengan sistem operasi komputer. Dalam _shell_ pengguna dapat memberikan perintah-perintah untuk melihat informasi sistem, mengelola file, bahkan mengatur kinerja dari komputer. Ini adalah antarmuka berbasis teks yang dikenal dengan TUI (_Text User Interface_). Shell beroperasi dalam terminal atau konsol. Program shell sendiri beragam yang memiliki cara penggunaan masing-masing. Yang digunakan saat ini adalah **BASH**, yaitu salah satu program _shell_ yang lengkap dan umum digunakan. Dalam shell kita bisa memanggil perintah bawaan, program yang tersedia dalam sistem operasi, juga program-program yang dibuat oleh user. Urutan perintah-perintah dalam _shell_ dapat dituliskan dalam sebuah file yang dikenal dengan _shell-script_. Cara ini sangat membantu melakukan otomatisasi proses sehingga pekerjaan komputasi menjadi lebih efisien.

Menggunakan Jupyter, _shell_ dapat dibuka melalui ikon tombol terminal. Setelah klik tombol tersebut akan terbuka jendela seperti berikut,

![](term.png)

Blok kursor hitam menunjukkan _prompt_ yaitu tempat untuk memberikan perintah. Perintah dieksekusi per baris dengan jalan menekan tombol enter. Perintah ini dikenal dengan baris perintah atau _command line_. Teks yang berada di belakang blok kursor adalah informasi _prompt_ yang formatnya dapat diatur. Pengaturan baku dari _shell_ dilakukan dalam sebuah file configurasi bernama .bashrc tepat di bawah direktori rumah (_home directory_). File ini adalah sebut _shell script_.

Berikut beberapa aspek dalam _shell_ yang perlu diketahui,

1. _Shell_ dapat memiliki variabel yang diberi istilah _environment variable_ atau variabek lingkungan. Variabel ini dapat digunakan untuk menyimpan data teks, bilangan, atau array. Sebuah variabel akan berlaku hanya pada shell yang aktif. Agar variabel dikenal oleh program yang dijalankan melalui baris perintah harus dideklarasikan menggunakan perintah **export**.
2. _Shell_ memiliki elemen-elemen pemrograman seperti pengujian kondisi `if-then-else` dan pengulangan, seperti `for` dan `while`
3. _Shell_ mengenal blok program dan fungsi

Walaupun _shell_ dapat digunakan sebagai bahasa pemrograman, namun kegunaan utamanya adalah untuk melakukan administrasi sistem dan pengelolaan berkas. Penamaan dan pengalamatan file atau direktori mengikuti aturan berikut,

- Nama file dan direktori sensitif terhadap bentuk huruf. Huruf kapital dianggap berbeda dengan huruf kecil.
- Penggunaan tanda-tanda, walaupun memungkinkan, sebaiknya dihindari untuk mempermudah pengalamatan, terutama dalam pemrograman. Spasi sebaiknya diganti dengan garis bawah (_underscore_)
- Pemisah antara nama direktori dengan direktori dibawahnya menggunakan garis miring `/`
- Tanda garis miring di awal direktori artinya direktori akar (_root_), ini digunakan untuk pengalamatan absolut.
- Jika garis miring tidak digunakan artinya sub-direktori atau file yang berada di direktori kerja yang aktif.
- Dalam pengalamatan direktori;
  - `.` artinya _current directory_
  - `..` artinya _upper directory_
  - `~` artinya _home directory_
- Tanda titik yang digunakan di awal nama bermakna _hidden file_

Berikut adalah beberapa perintah dasar yang perlu diketahui,

- Melihat isi dari direktori: `ls`
- Pindah direktori: `cd nama_direktori` \
  `nama_direktori` mengikuti aturan penamaan direktori.
- membuat direktori: `mkdir nama_direktori` \
  Aturan penamaan sama dengan diatas
- menghapus file: `rm nama_file` atau `rm -r nama_direktori`
- Melihat isi file (teks): `cat nama_file`
- Menampilkan teks atau variabel di layar: `echo teks` atau `echo $NAMA_VARIABEL`
- Membersihkan layar: `clear`

Masih sangat banyak perintah-perintah _shell_ yang berguna, anda sangat disarankan untuk mempelajari tentang **BASH** lebih lanjut dari [dokumen manualnya.](https://www.gnu.org/savannah-checkouts/gnu/bash/manual/bash.html)

Variabel baku dalam **BASH** shell,

- `$HOME` direktori rumah dari pengguna. Variabel ini diberi nama alias `~`
- `$PS1` teks yang diberikan sebagai penanda _prompt_
- `$PWD` alamat direktori yang digunakan
- `$PATH` penjejakan program eksekusi

Selain variabel `$HOME` dan `$PWD`, variabel dapat diubah sesuai dengan keinginan pengguna. Tanda dolar (`$`) digunakan untuk mengakses variabel yang sudah ada. Tanda ini tidak digunakan saat mendeklarasikan variabel, contoh,
```bash
echo $HOME
A='text variabel'
N=100
export NAMAKU='mhs geofisika'
echo $A $N
```
Dalam contoh ini isi variabel `$HOME` diakses ketika memanggil perintah `echo`. Ketika mendeklarasikan variabel seperti pada contoh `A='text variable'` tanda dolar tidak boleh dipergunakan.

Untuk kenyamanan penggunaan sebaiknya dilakukan konfigurasi terlebih dahulu dengan cara mengedit berkas `.bashrc` di direktori _home_. Yang biasa diatur dalam file ini adalah `$PATH` dan `$PS1`. Anda juga dapat membuat nama-nama alias untuk perintah perintah yang sering digunakan sehingga memudahkan pengetikan. 

Variabel `$PATH` sangat penting untuk mengarahkan koleksi program perintah, sehingga sistem akan mencari program yang dipanggil di direktori yang terdaftar dalam path. User dapat memasukan banyak direktori pencarian seperti berikut,
```bash
export PATH=$HOME/bin:/opt/koleksi/bin:$PATH
```
Dalam contoh ini `$HOME/bin` dan `/opt/koleksi/bin` dimasukan sebagai alamat pencarian program, sehingga program-program eksekusi yang ada di sana dapat dipanggil langsung dari _prompt_.

Untuk mengubah teks informasi _prompt_ perlu mengetahui terlebih dulu simbol-simbol yang dapat digunakan. Contoh yang sederhana adalah sebagai berikut,
```bash
export PS1='Latihan: \u@\h [\d]\n-> '
```
akan menghasilkan _prompt_ seperti berikut,
```
Latihan: mhs@numeric [Sat Feb 21]
-> 
```

Anda dapat pula mendesain tampilan _prompt_ misalnya melalui halaman [bash prompt generator](https://bash-prompt-generator.org/). Jika diinginkan pengaturan yang lebih modern dapat menggunakan [starship](https://starship.rs/), yang dapat diinstall dalam sistem kerja anda menggunakan perintah,
```bash
sudo apk add starship
```

Opsi `h` dalam perintah untuk menampilkan besar file dalam format yang dimengerti manusia, menjadi `ls -lh`. Berikut adalah contoh keluaran setelah melakukan perintah diatas. 

```
((numeric) ) numeric:/workdata$ ls -lh
total 2M     
-rw-r--r--    1 mhs      mhs        10.5K Feb 21 06:15 00_Workspace.ipynb
-rw-r--r--    1 mhs      mhs        24.0K Feb 20 08:34 02_Matriks.ipynb
-rw-r--r--    1 mhs      mhs        18.5K Feb 20 08:34 03_Eliminasi.ipynb
-rw-r--r--    1 mhs      mhs       369.8K Feb 20 08:35 04_Interpolasi.ipynb
-rw-r--r--    1 mhs      mhs       124.4K Feb 16 02:19 05_Extrapolasi.ipynb
-rw-r--r--    1 mhs      mhs       185.6K Feb 16 03:03 06_Akar.ipynb
-rw-r--r--    1 mhs      mhs       130.7K Feb 20 08:39 07_Turunan.ipynb
-rw-r--r--    1 mhs      mhs        60.0K Feb 20 08:29 08_Integral.ipynb
-rw-r--r--    1 mhs      mhs        18.8K Feb 12 15:47 PoliFit.ipynb
-rw-r--r--    1 mhs      mhs          157 Feb 21 03:08 compose.yml
-rw-r--r--    1 mhs      mhs        23.9K Feb 13 00:57 curve_fit.ipynb
-rw-r--r--    1 mhs      mhs       391.4K Feb 20 11:03 dockerweb.png
-rw-r--r--    1 mhs      mhs        50.8K Feb 21 00:46 jupy.png
-rw-r--r--    1 mhs      mhs         1.8K Feb 12 03:03 regresi.ipynb
-rw-r--r--    1 mhs      mhs        64.5K Feb 15 03:42 rootwhale.png
-rw-r--r--    1 mhs      mhs         5.8K Feb 21 04:05 term.png
-rw-r--r--    1 mhs      mhs        89.2K Feb 15 03:25 whale.png
drwxr-xr-x    1 mhs      mhs           58 Feb 21 03:08 work
```

Perhatikan kolom pertama dari daftar file yang memiliki format seperti,
`-rw-r--r--`. Huruf-huruf ini memiliki makna sebagai berikut,

- Huruf pertama menunjukkan jenis file, `-` adalah file biasa sedangkan `d` adalah direktori.
- Sembilan huruf berikutnya adalah _file-permission_ dibagi tiga kolom, kolom pertama untuk pemilik (`u`, _user_), kedua untuk yang memiliki grup sama dengan pemilik (`g`, _group_), dan yang ketiga untuk semua user dalam sistem (`o`, _other_).
- yang isinya bisa huruf `rwx` atau tanda minus `---`. `r` menandakan file bisa dibaca, `w` bisa ditulis, dan `x` bisa dijalankan (_excecutable_). Untuk direktori `x` memiliki makna lain, yaitu bisa dilihat (_browseable_).

Untuk mengubah perijinan digunakan perintah `chmod` yang penggunaannya dapat dilihat di [manual](https://man7.org/linux/man-pages/man1/chmod.1.html).

## Pemrograman

Pemrograman adalah pekerjaan implementasi dari algoritma ke dalam program dalam bahasa yang dimengerti komputer.
Program ini dikerjakan baris demi baris oleh komputer. Program dituliskan dalam kode sumber (_source code_) kemudian dijalankan oleh komputer (_eksekusi_). Eksekusi program dapat dilakukan melalui dua cara,

**Kompilasi**.

Kode sumber diubah bentuk menjadi bahasa mesin kemudian dimuat ke dalam memori dan dijalankan. Proses kompilasi akan mengubah seluruh badan program menjadi kode yang dimengerti mesin. Sehingga tidak diperkenankan ada kesalahan penulisan kode sumber. Perangkat lunak yang digunakan untuk melakukan kompilasi disebut dengan _compiler_. Proses kompilasi digambarkan pada _flowchart berikut_,

![kompilasi program](kompilasi.png)

Contoh dari bahasa yang menggunakan kompiler adalah, `C/C++`, `Go`, dan `Rust`. Proses kompilasi dapat dijelaskan dalam diagram alir dibawah. _Compiler_ hanya melakukan konversi dari kode sumber menjadi kode mesin yang disebut dengan _op-code_, tetapi tidak melakukan mengujian benar dan salahnya algoritma program. Selama kode sumber tidak memiliki kesalahan sintaks, maka file biner akan dibuat. Kesalahan logika dapat diketahui pada saat program dijalankan (eksekusi). Pencarian kesalahan algoritma atau logika program dapat dilakukan dengan mengamati hasil program (luaran), dan dapat dilakukan dengan efektif menggunakan _debugger_.

![proses kompilasi](proses_kompilasi.png)

**Interpretasi**

Kode sumber dijalankan baris perbaris secara langsung. Perangkat lunak yang melakukan interpretasi disebut dengan _interpreter_, bertugas untuk menterjemahkan baris perintah dan mengirim ke mesin komputer untuk dieksekusi. Kode sumber dapat dijalankan tanpa melalui proses kompilasi, sehingga jika ada kesalahan penulisan di satu baris, program masih bisa berjalan sebelum baris yang salah ditemukan. Interpretasi diilustrasikan dalam _flowchart_ berikut,

Dalam ilustrasi diatas blok di dalam loop, pengambilan dan penjalanan program dilakukan oleh interpreter, bukan oleh pemrogram. Yang harus diperhatikan adalah, program dibaca dan dijalankan baris perbaris. Contoh bahasa pemrograman yang menggunakan teknin interpretasi adalah semua _shellscript_, Python, R, dan Julia.

## Pemrograman menggunakan _Shellscript_

**Inisialisasi dan mengakses veriabel**

```bash
#!/bin/bash

data=10
echo Tanggal hari ini $(date)

echo berikut adalah penggunaan for loop dalam shellscript
for i in 2 3 4 5; do
  echo "$data * $i =" $(( $data + $i ))
done

echo --------------
echo program di jalankan di komputer $HOSTNAME oleh user $USER
```

Baris pertama pada contoh program diatas adalah menunjuk untuk interpreter skrip yang digunakan. Tanda pagar `#` umumnya digunakan untuk menyatakan baris komentar, namun jika diletakan pada paris pertama dan diikuti oleh tanda sertu `!`, memiliki arti penentu program interpreter untuk skrip, dalam hal ini `bash` yang dituliskan dalam alamat penjejakan yang lengkap. Agar file skrip dapat dijalankan langsung dari baris perintah, maka modus eksekusi dari _file permission_ harus diaktifkan. Misalnya file diatas disimpan dengan nama `script.sh`, perintah berikut digunakan untuk menjadikan file tersebut _executable_.
```bash
chmod +x script.sh
```
Kemudian langsung dijalankan dengan,
```bash
./script.sh
```

Pemrograman dengan _shellskrip_ seringkali membantu mengerjakan perkerjaan-pekerjaan administrasi rutin menjadi lebih mudah. Dalam skrip anda dapat memanggil perintah-perintah yang biasa dilakukan di baris perintah. Kebanyakan proses-proses pada baris perintah dikerjakan menggunakan program lain. Contohnya penggunaan perintah `date` pada program diatas untuk mengambil tanggal dan waktu dari sistem.

Contoh berikut adalah contoh untuk manajemen file.

```bash
#!/bin/bash
# instalasi program fortune jika belum ada
sudo apk add fortune
DATADIR=textdata
BCKFILE=data.tgz
mkdir -p $DATADIR
cd $DATADIR

for (( i=0; i<10; i++ )); do
  fortune > text-$i.txt
done

echo Daftar file yang dibuat
echo -------------------------------------
ls -lh
echo -------------------------------------
echo

for i in $(ls text-*); do
  echo "isi file $i >>>"
  cat $i
  echo
done

# melakukan backup data
cd ..
tar czv $DATADIR > $BCKFILE
echo
echo data sudah dibackup ke $BCKFILE
echo isi dari file backup:
tar tzvf $BCKFILE
echo untuk mengekstrak data dari file $BCKFILE: tar xzvf $BCKFILE
```


## Pemrograman menggunakan C

C adalah bahasa pemrograman yang memerlukan proses kompilasi. Program kompiler dalam linux adalah `gcc`. 

```c
#include <stdio.h>
float data=1.23;
float c;

int main() {
    printf("Program pertama dalam C\n");
    printf("Ini adalah variabel yang dideklarasikan data=%f\n", data);
    // berikut adalah contoh loop dalam bahasa C
    for (int i=0; i<10; i++) {
        printf("nilai iterator i=%d\n",i);
        c=(float)i+data;
        printf("operasi bilangan %d+%0.2f=%0.3f\n", i, data, c);
    } 
    return 0;
}
```

Untuk menjalankan program dilakukan tahap-tahap berikut,

1. Buat dan edit program diatas dalam file teks dengan akhiran `.c` misalnya `lati01.c`
2. Kompilasi program dengan perintah \
   `gcc lati01.c -o lati01`
3. Lihat jika file program `lati01` sudah ada, menggunakan `ls -l`. Perhatikan perijinan filenya.
4. Jalankan program dengan memanggil `./lati01` di baris perintah. Perhatikan bahwa eksekusi program harus dari direktori tempat file program tersebut berada.

Program yang ditulis dalam bahasa C dijalankan oleh komputer lebih cepat daripada skrip, karena komputer langsung membaca _op-code_ dari memori. Untuk pekerjaan-pekerjaan yang sangat kompleks, bahasa ini digunakan. Misalnya untuk membuat sebuah sistem operasi.

## Pemrograman menggunakan Python

Python adalah salah satu bahasa scrip (_scripting language_), namun berbeda dengan _bash script_, python secara khusus didesain untuk menjalankan algoritma-algoritma yang kompleks, seperti penyelesaian masalah matematika. Python populer dalam aplikasi yang berkaitan dengan numerik dan data. Ia telah dilengkapi dengan hampir semua jenis algoritma yang diperlukan untuk pekerjaan analisis numerik. Saat ini Python menjadi bahasa yang banyak digunakan untuk _machine learning_ dan _data science_. Pustaka-pustaka dasarnya ditulis dalam bahasa C/C++, sehingga untuk pekerjaan tertentu memiliki kecepatan proses yang baik.

Berbeda dengan bahasa-bahasa pemrograman lain, dalam Python blok ditandai dengan banyaknya indentasi, spasi atau tabulasi, yang digunakan. Berikut contoh dari program dalam Python.

```python
#!/usr/bin/env python

# deklarasi data
data=1.23
darr=[1, 3, 5, 2, 5]

print("Program dalam Python")
print(f'Isi data bilangan nyata: {data:0.2f}')
print(f'Data array:\n{darr}')

for i in darr:
    print(f'operasi data {i}+{data}={i+data}')
```

Program ini dapat dijalankan tanpa melalui kompilasi. Baris pertama merupakan penunjuk bagi interpreter yang digunakan. Karena Python biasa menggunakan lingkungan program khusus yang disebut _Python environment_, maka penggunaan program `/usr/bin/env` penting untuk mendapat distribusi Python yang benar. Prosedur untuk menjadikan file ini _executable_ dan cara menjalankannya dilakukan sama dengan pada _shellscript_. 

## Format data dalam dalam file

Data disimpan dalam media penyimpanan dapat berupa file biner dan file teks. File biner tidak dapat secara langsung dibaca isinya, sedangkan file teks dapat dilihat dan disunting secara langsung di terminal atau menggunakan editor teks.

**Plain text**

_Plain text_ atau teks mentah adalah teks tanpa format. Data teks ini biasanya diorganisasikan dengan baris. Harus diingat bahwa karakter yang digunakan untuk menandakan akhir baris berbeda bergantung sistem operasi yang digunakan. Dalam bahasa C dan Python menggunakan karakter `Line Feed (LF)` yaitu karakter `0x10`. Didalam program karakter ini ditulis sebagai _escape character_ `\n` dalam variabel string atau karakter.

**File teks terstruktur**

File data terpisahkan dengan koma, Comma Separated Values (CSV).

Contoh,

```csv
Nomor,Nama,Jenis,Harga
1,Tomat,buah,10000
2,Kangkung,sayur,2000
3,Apel,buah,15000
4,Jeruk,buah,5000
6,Bayam,sayur,3000
```

File data dalam format JSON.

```json
{
    "info": "tabel harga sayur dan buah berdasarkan jenis",
    "sayur": [
        {"nama": "kangkung", "harga": 2000},
        {"nama": "Bayam", "harga": 3000}
    ],
    "buah": [
        {"name": "Tomat", "harga": 10000},
        {"name": "Apel", "harga": 15000},
        {"name": "Jeruk", "harga": 5000}
    ]
}
```
